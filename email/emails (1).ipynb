Classify the email using the binary classification method. Email Spam detection has two
states: a) Normal State – Not Spam, b) Abnormal State – Spam. Use K-Nearest Neighbors and
Support Vector Machine for classification. Analyze their performance.
Dataset link: The emails.csv dataset on the Kaggle
https://www.kaggle.com/datasets/balaka18/email-spam-classification-dataset-csv


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, accuracy_score, precision_score, recall_score, plot_precision_recall_curve, plot_roc_curve
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# Step 1: Load the dataset
df = pd.read_csv('emails.csv')

# Step 2: Check for missing values and clean data
print(df.isnull().sum())  # To check for any missing values
df.dropna(how='any', inplace=True)  # Drop rows with missing values

# Step 3: Split the dataset into features (X) and target (y)
X = df.iloc[:, 1:-1].values  # All columns except the first (ID) and last (target)
y = df.iloc[:, -1].values    # The last column is the target (spam or not spam)

# Step 4: Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)

# Function to report model performance
def report(classifier):
    y_pred = classifier.predict(x_test)
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    display = ConfusionMatrixDisplay(cm)
    display.plot()
    plt.show()
    
    # Print accuracy, precision, recall
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
    print(f"Precision Score: {precision_score(y_test, y_pred):.2f}")
    print(f"Recall Score: {recall_score(y_test, y_pred):.2f}")
    
    # Precision-Recall Curve
    plot_precision_recall_curve(classifier, x_test, y_test)
    plt.show()
    
    # ROC Curve
    plot_roc_curve(classifier, x_test, y_test)
    plt.show()

# Step 5: Apply K-Nearest Neighbors (KNN)
kNN = KNeighborsClassifier(n_neighbors=10)
kNN.fit(x_train, y_train)
print("K-Nearest Neighbors Performance:")
report(kNN)

# Step 6: Apply Support Vector Machine (SVM)
svm = SVC(gamma='auto', random_state=10)
svm.fit(x_train, y_train)
print("Support Vector Machine Performance:")
report(svm)
```
